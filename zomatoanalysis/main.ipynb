```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
# preprocess data
 
df.rename(columns = {'approx_cost(for two people)':'price', 'listed_in(type)':'type'}, inplace=True)

def handleRate(value):
    return(float(str(value).split('/')[0]))

df['rate'] = df['rate'].apply(handleRate) 
```

```python
# Bar plot distribution of type of restaurants
sns.countplot(x=df['type'], palette='Set1')
plt.xlabel("Type of restaurant")
plt.title("Bar Plot Distribution of Type Of Restaurants")
```

```python
plt.plot(df.groupby('type')['votes'].sum(), c='blue', marker='p')
plt.xlabel("Type of Restaurant", c='green', size = 15)
plt.ylabel("Votes", c='green', size = 15)
plt.title("Line Plot of Votes against Type of Restaurant")
```

```python
sns.countplot(x=df['online_order'], color='red')
plt.title("Distribution of online orders")
```

```python
plt.hist(df['rate'], bins=5)
plt.title("Ratings Distribution")
plt.xlabel("Ratings")
plt.ylabel("Count")
```

```python
sns.countplot(data=df, x='price', palette='pastel')
plt.title("Distribution of Approximate Cost (for two people)")
plt.xlabel("Approx. Cost")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate', data = df)
```

```python
sns.heatmap(df.pivot_table(index='type',columns='online_order', aggfunc='size', fill_value=0), annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap of Type of Restaurants against Online Orders")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
```
